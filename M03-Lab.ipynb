```python
# Name: Gabriel Wilson
# Filename: M03-Lab.ipynb
# Description: Collect car model information from the user and store it in a Vehicle class.

# base class
class Vehicle:
    def __init__(self, type):
        self.type = type

    def __str__(self) -> str:
        return f"Vehicle type: {self.type}"

# child class
class Automobile(Vehicle):
    def __init__(self, year, make, model, doors, roof):
        super().__init__("car")
        self.year = year
        self.make = make
        self.model = model
        self.doors = doors
        self.roof = roof

    def __str__(self) -> str:
        return super().__str__() + f"\nYear: {self.year}\nMake: {self.make}\nModel: {self.model}\nNumber of doors: {self.doors}\nType of roof: {self.roof}"

# gets and validates user input, then returns an integer value
def int_input(prompt) -> int:
    while True:
        strInput = input(prompt)

        try:
            value = int(strInput)
            return value
        except:
            print("invalid. input must be a number")


def main():
    print("please input car information below:")

    # get year
    year = int_input("year: ")

    # get make
    make = input("make: ")

    # get model
    model = input("model: ")

    # get number of doors
    while True:
        doors = int_input("number of doors (2 or 4): ")
        if doors == 2 or doors == 4:
            break
        else:
            print("invalid number of doors")

    # get roof type
    while True:
        roof = input("roof type ('solid' or 'sun roof'): ")
        if roof.lower() == 'solid' or roof.lower() == 'sun roof':
            break
        else:
            print("invalid roof type")

    # create object
    car = Automobile(year, make, model, doors, roof)

    # print results
    print(car)


if __name__ == "__main__":
    main()
```
